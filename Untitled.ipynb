```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import os

def fingerprint_processing_pipeline(img_path):
    # ---------- STEP 1: Load Image ----------
    if not os.path.exists(img_path):
        raise FileNotFoundError(f"Image not found at path: {img_path}")

    original_img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if original_img is None:
        raise ValueError("Image couldn't be loaded. Check the file format and path.")

    # ---------- STEP 2: Preprocessing ----------
    blurred_img = cv2.GaussianBlur(original_img, (5, 5), 0)

    def normalize(img):
        return cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

    normalized_img = normalize(blurred_img)

    # ---------- STEP 3: Segmentation with ROI padding ----------
    _, binary_img = cv2.threshold(normalized_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    binary_img = 255 - binary_img  # Invert for ridges white

    contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        raise ValueError("No contours found in fingerprint image.")

    x, y, w, h = cv2.boundingRect(contours[0])
    pad = 10
    x = max(0, x - pad)
    y = max(0, y - pad)
    w = min(normalized_img.shape[1] - x, w + 2 * pad)
    h = min(normalized_img.shape[0] - y, h + 2 * pad)

    segmented_img = normalized_img[y:y + h, x:x + w]

    # ---------- STEP 4: Edge Detection ----------
    def apply_sobel(img):
        grad_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
        grad_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
        edge_mag = cv2.magnitude(grad_x, grad_y)
        return np.uint8(np.clip(edge_mag, 0, 255))

    edge_img = apply_sobel(segmented_img)

    # ---------- STEP 5: Binarization and Thinning ----------
    _, binarized_img = cv2.threshold(segmented_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    binarized_img = 255 - binarized_img  # Ridges white

    def zhang_suen_thinning(image, max_iterations=100):
        image = image.copy() // 255
        iteration = 0
        while iteration < max_iterations:
            marker = np.zeros_like(image)
            changing = False
            for step in [0, 1]:
                for i in range(1, image.shape[0] - 1):
                    for j in range(1, image.shape[1] - 1):
                        P = [image[i, j], image[i - 1, j], image[i - 1, j + 1], image[i, j + 1], image[i + 1, j + 1],
                             image[i + 1, j], image[i + 1, j - 1], image[i, j - 1], image[i - 1, j - 1]]
                        if (P[0] == 1 and
                                2 <= sum(P[1:]) <= 6 and
                                sum((P[1] == 0 and P[2] == 1,
                                     P[2] == 0 and P[3] == 1,
                                     P[3] == 0 and P[4] == 1,
                                     P[4] == 0 and P[5] == 1,
                                     P[5] == 0 and P[6] == 1,
                                     P[6] == 0 and P[7] == 1,
                                     P[7] == 0 and P[8] == 1,
                                     P[8] == 0 and P[1] == 1)) == 1 and
                                (P[1] * P[3] * P[5] if step == 0 else P[1] * P[5] * P[7]) == 0 and
                                (P[3] * P[5] * P[7] if step == 0 else P[1] * P[3] * P[7]) == 0):
                            marker[i, j] = 1
                            changing = True
                image = image & ~marker
                marker[:] = 0
            if not changing:
                break
            iteration += 1
        return (image * 255).astype(np.uint8)

    # Resize image for thinning speed-up (optional)
    small_img = cv2.resize(binarized_img, (0, 0), fx=0.5, fy=0.5)
    skeleton_img = zhang_suen_thinning(small_img)

    # ---------- STEP 6: Minutiae Extraction ----------
    def get_minutiae(skeleton_img):
        endings, bifurcations = [], []
        img = skeleton_img // 255
        for i in range(1, img.shape[0] - 1):
            for j in range(1, img.shape[1] - 1):
                if img[i, j] == 1:
                    neighbors = [img[i - 1, j], img[i - 1, j + 1], img[i, j + 1], img[i + 1, j + 1],
                                 img[i + 1, j], img[i + 1, j - 1], img[i, j - 1], img[i - 1, j - 1]]
                    cn = sum(abs(int(neighbors[k]) - int(neighbors[(k + 1) % 8])) for k in range(8)) // 2
                    if cn == 1:
                        endings.append((j, i))
                    elif cn == 3:
                        bifurcations.append((j, i))
        return endings, bifurcations

    endings, bifurcations = get_minutiae(skeleton_img)

    # ---------- STEP 7: Clean Minutiae ----------
    def clean_minutiae(minutiae_points, image_shape, border_margin=10, isolation_distance=15):
        h, w = image_shape
        cleaned = [(x, y) for x, y in minutiae_points
                   if border_margin < x < w - border_margin and border_margin < y < h - border_margin]

        final = []
        for i, (x1, y1) in enumerate(cleaned):
            close = sum(np.hypot(x1 - x2, y1 - y2) < isolation_distance
                        for j, (x2, y2) in enumerate(cleaned) if i != j)
            if close > 0:
                final.append((x1, y1))
        return final

    cleaned_endings = clean_minutiae(endings, skeleton_img.shape)
    cleaned_bifurcations = clean_minutiae(bifurcations, skeleton_img.shape)

    print(f"Original Endings: {len(endings)}, Cleaned Endings: {len(cleaned_endings)}")
    print(f"Original Bifurcations: {len(bifurcations)}, Cleaned Bifurcations: {len(cleaned_bifurcations)}")

    # ---------- STEP 8: Graph-Based Ridge Structure ----------
    G = nx.Graph()
    for x, y in cleaned_endings:
        G.add_node((x, y), type='ending')
    for x, y in cleaned_bifurcations:
        G.add_node((x, y), type='bifurcation')

    distance_threshold = 20
    all_points = list(G.nodes)
    for i, (x1, y1) in enumerate(all_points):
        for j in range(i + 1, len(all_points)):
            x2, y2 = all_points[j]
            dist = np.hypot(x1 - x2, y1 - y2)
            if dist <= distance_threshold:
                G.add_edge((x1, y1), (x2, y2), weight=dist)

    print(f"Graph created with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")

    plt.figure(figsize=(8, 8))
    pos = {node: node for node in G.nodes}
    nx.draw(G, pos, node_color='lightblue', node_size=10, edge_color='gray', width=0.2)
    plt.title("Minutiae Graph Visualization")
    plt.axis('equal')
    plt.show()
 

   # ---------- STEP 9: Interactive Visualization ----------
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.imshow(skeleton_img, cmap='gray')
    
    if cleaned_endings:
        ax.scatter(*zip(*cleaned_endings), c='red', marker='o', label='Endings')
    if cleaned_bifurcations:
        ax.scatter(*zip(*cleaned_bifurcations), c='blue', marker='x', label='Bifurcations')
    
    ax.legend()
    ax.set_title("Click a minutiae point to see info")
    ax.axis('off')
    
    # Combine minutiae points
    minutiae_coords = endings + bifurcations
    
    # Single, clean on_click function
    def on_click(event):
        if event.inaxes != ax:
            return
    
        x_click, y_click = event.xdata, event.ydata
    
        min_dist = float('inf')
        closest_point = None
    
        for x, y in minutiae_coords:
            dist = ((x - x_click) ** 2 + (y - y_click) ** 2) ** 0.5
            if dist < min_dist:
                min_dist = dist
                closest_point = (x, y)
    
        if min_dist < 10:
            node_type = G.nodes.get(closest_point, {}).get('type', 'Unknown')
            neighbors = list(G.neighbors(closest_point)) if closest_point in G else []
            print(f"Clicked Minutiae: ({closest_point[0]}, {closest_point[1]})")
            print(f"Type: {node_type}")
            print(f"Connected to: {neighbors}")
        else:
            print("No minutiae point near click.")
    
    # Attach the event
    fig.canvas.mpl_connect('button_press_event', on_click)
    
    plt.show()
    
        
    # ---------- STEP 10: Skeleton Quality Check ----------
    fig, axs = plt.subplots(2, 3, figsize=(15, 10))
    axs = axs.flatten()
    axs[0].imshow(original_img, cmap='gray')
    axs[0].set_title('Original Image')
    axs[1].imshow(blurred_img, cmap='gray')
    axs[1].set_title('Blurred Image')
    axs[2].imshow(normalized_img, cmap='gray')
    axs[2].set_title('Normalized Image')
    axs[3].imshow(segmented_img, cmap='gray')
    axs[3].set_title('Segmented Image')
    axs[4].imshow(edge_img, cmap='gray')
    axs[4].set_title('Edge Detection')
    axs[5].imshow(skeleton_img, cmap='gray')
    axs[5].set_title('Skeleton Image')
    for ax in axs:
     ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
fingerprint_processing_pipeline("input/fingerprint1.jpg")
```
